### Data Cleaning Steps

```python
import pandas as pd
df = pd.read_csv("train.csv")
df.head(10)
```

```python
#Get column names
column_names = df.columns
print(column_names)
#Get column data types
df.dtypes
#Also check if the column is unique
for i in column_names:
    print("{} is unique : {}".format(i,df[i].is_unique))
```

```python
# Check the index values
df.index.values
```

```python
# Check if a certain index exists
0 in df.index.values
```

```python
# If index does not exist
df.set_index("employee_id",inplace=True)
```

```python
df
```

```python
#Create list comprehension of the columns you want to lose
columns_to_drop = [column_names[i] for i in [8,9,10]]
```

```python
#Drop unwanted columns
df.drop(columns_to_drop, inplace=True, axis=1)
```

```python
df
```

```python
#fill NaN with ' '
df['department'] = df['department'].fillna(' ')
df
```

```python
#Fill NaN with 99
df['education'] = df['education'].fillna(99)
df
```

```python
#Fill NaN with the mean of the column
df['age'] = df['age'].fillna(df['age'].mean())
df
```
